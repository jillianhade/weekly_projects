### How do I get rid of multi-columns? 

There's not really a good way. If you want to squish all the names together, the following will do it (make a new dataframe out of your old one, but with `.to_records()` on there)

```python
pd.DataFrame(test_data.to_records())
```

That makes for kind of ugly column names though!

The other option is to drop one level of those columns like below, but that will require remember what order things are in and renaming columns. Kind of a pain, but another option

```python
test_data.columns = test_data.columns.droplevel(0)
test_data
```

```python
def timeSplit(string):
    temp = string.split(":")
    minutes = int(temp[0])
    seconds = int(temp[1])
    return minutes*60 + seconds
```

```python
df['time_in_seconds'] = df['time'].apply(timeSplit)
```

```python
def year_extractor(string):
    temp = string.split("-")
    return int(temp[0])

def month_extractor(string):
    temp = string.split("-")
    return int(temp[1])

def day_extractor(string):
    temp = string.split("-")
    return int(temp[2])

year_extractor('2000-09-23')
```

```python
month_extractor('2000-09-23')
```

```python
day_extractor('2000-09-23')
```

```python
df['month_entered_billboard'] = df['date.entered'].apply(month_extractor)
```
